```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification, AutoConfig, pipeline
from datasets import load_dataset

# Load the IMDb dataset
ds = load_dataset("jahjinx/IMDb_movie_reviews")

model_name = "JiaqiLee/imdb-finetuned-bert-base-uncased"
token = "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"  # Replace with your Hugging Face token

# Load the tokenizer with max_length to avoid exceeding token limit
tokenizer = AutoTokenizer.from_pretrained(model_name, use_auth_token=token)

# Load the configuration
config = AutoConfig.from_pretrained(model_name, use_auth_token=token)

# Ensure the config has id2label
config.id2label = {0: "Negative", 1: "Positive"}  # Manually adding labels

# Load the model as a Sequence Classification model
model = AutoModelForSequenceClassification.from_pretrained(model_name, config=config, use_auth_token=token, trust_remote_code=True)

# Set the maximum length for truncation
max_length = 512  # Adjust this based on the model's limit

# Create the text classification pipeline with truncation and max_length enabled
classifier = pipeline("text-classification", model=model, tokenizer=tokenizer, device=-1, truncation=True, padding=True, max_length=max_length)

# Test with some data from the IMDb dataset, reducing to 500 samples for quick testing
test_texts = ds["test"]["text"][:500]

# Get predictions directly with raw text (classifier handles tokenization and truncation)
predictions = classifier(test_texts)

# Print predictions
for i in range(5):  # Print the first 5 predictions
    print(f"Text: {test_texts[i]}")
    print(f"Predicted Label: {predictions[i]['label']}")
```

```python
from sklearn.metrics import accuracy_score, precision_recall_fscore_support

# Extract true labels (assume labels are already 0 for Negative, 1 for Positive)
test_labels = ds["test"]["label"][:500]  # Extract corresponding labels

# Extract predicted labels as binary values (0 or 1)
predicted_labels = [1 if pred["label"] == "Positive" else 0 for pred in predictions]

# Compute standard evaluation metrics
accuracy = accuracy_score(test_labels, predicted_labels)
precision, recall, f1, _ = precision_recall_fscore_support(test_labels, predicted_labels, average="binary")

# Print metrics
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
```

```python
import sacrebleu
pred_texts = [pred['label'] for pred in predictions]  # Predicted labels as texts
true_texts = test_texts  # Actual text from the dataset (for reference)
bleu_score = sacrebleu.corpus_bleu(pred_texts, [true_texts]).score
```

```python
from rouge_score import rouge_scorer
# Compute ROUGE Scores
scorer = rouge_scorer.RougeScorer(["rouge1", "rouge2", "rougeL"], use_stemmer=True)
rouge_scores = [scorer.score(ref, pred) for ref, pred in zip(true_texts, pred_texts)]

# Aggregate ROUGE scores
rouge_1 = sum(score["rouge1"].fmeasure for score in rouge_scores) / len(rouge_scores)
rouge_2 = sum(score["rouge2"].fmeasure for score in rouge_scores) / len(rouge_scores)
rouge_L = sum(score["rougeL"].fmeasure for score in rouge_scores) / len(rouge_scores)
```
